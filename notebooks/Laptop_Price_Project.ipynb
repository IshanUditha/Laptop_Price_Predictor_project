```python
def add_company(input):
    if input =="Samsung" or input =="Razer" or input =="Mediacom" or input =="Microsoft" or input =="Xiaomi" or input =="Vero" or input =="Chuwi" or input =="Google" or input =="Fujitsu" or input =="LG" or input == "Huawei":
        return "Other"
    else:
        return input

data["Company"] = data["Company"].apply(add_company)
```

```python
def set_processor(name):
    if name == "Intel Core i3" or name == "Intel Core i5" or name == "Intel Core i7":
        return name
    elif name.split()[0] == "AMD":
        return "AMD"
    else:
        return "Other"

data["Cpu_name"] =data["Cpu_name"].apply(set_processor) 
    
```

```python
def set_os(input):
    if input=="Windows 10" or input =="Windows 7" or input =="Windows 10 S":
        return "Windows"
    elif input =="macOS" or input =="Mac OS X":
        return "Mac"
    elif input == "Linux":
        return input
    else:
        return "Other"

data["OpSys"] = data["OpSys"].apply(set_os)
```

```python
data = data.drop(["laptop_ID","Product","Inches","ScreenResolution","Cpu","Gpu"],axis =1)
```

```python
data["Memory_value"]= data["Memory"].apply(lambda x:" ".join(x.split()[0:2]))
```

```python
data["Memory_value"].value_counts()
```

```python
def set_memory_value(value):
    
    if value.split()[1] == "SSD":
        return "SSD"
    elif value.split()[1] == "HDD":
        return "HDD"
    else:
        return "Other"

data["Memory_value"] = data["Memory_value"].apply(set_memory_value)
```

```python
data_new["Ram"] = data["Ram"]
data_new["Weight"] = data["Weight"]
data_new["Price_euros"] = data["Price_euros"]
data_new
```

```python
x =data_new.drop(["Price_euros"],axis =1)
y =data_new["Price_euros"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
x_train.shape
```

```python
def model_acc(model):
    model.fit(x_train,y_train)
    acc = model.score(x_test,y_test)
    print(str(model) + " -> " + str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
model_acc(lr)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
model_acc(lasso)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
model_acc(dt)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
model_acc(rf)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {"n_estimators": [10,50,100],
             "criterion": ["squared_error", "absolute_error", "poisson"]}
grid_object =GridSearchCV(estimator = rf,param_grid =parameters)
grid_fit =grid_object.fit(x_train,y_train)
best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(x_test,y_test)
```

```python
import pickle
with open("Laptop_price_predictor.pickle","wb") as file:
    pickle.dump(best_model,file)
    
```
